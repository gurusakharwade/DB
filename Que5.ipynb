```python
import pandas as pd
import seaborn as sns
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
df = pd.DataFrame(np.column_stack((iris.data, iris.target)), columns = iris.feature_names+['target'])
df
```

```python
df['type'] = df.target.replace(dict(enumerate(iris.target_names)))
df
```

```python
df.groupby(['type'])['sepal length (cm)'].describe()
```

```python
df.groupby(['type'])['sepal width (cm)'].quantile(q=0.50)
```
