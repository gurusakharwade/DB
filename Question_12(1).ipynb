```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
print(df.to_string())
```

```python
df.info()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# Filling the missing data in age with mean value 
mean_age = df['age'].mean()
df['age'].fillna(mean_age, inplace = True)
```

```python
df.isnull().sum()
```

Box plot for distribution of age w.r.t. each gender along with the information about whether they survived or not

```python
sns.set_style('darkgrid')
plt.figure(figsize=(20,15))
sns.boxplot(x = 'sex', y = 'age', hue = 'survived', data = df)
plt.show()
```
